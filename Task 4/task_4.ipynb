## Train Test Split

```python
y=data['Species'].values
x=data[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']].values
```

```python
# Split the data into training and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
# Fitting on train set
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
# Making predicition on test set
pred=dt.predict(x_test)
```

```python
# Accuracy on classifcation report
from sklearn.metrics import accuracy_score,classification_report
print('Accuracy is',accuracy_score(pred,y_test))
print('\n')
print(classification_report(pred,y_test))
```

```python
# Importing required libraries for decision tree model
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
feature_column=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
```

```python
# Fitting the required functions to built decision tree algorithm
plt.figure(figsize=(30,20))
tree.plot_tree(dt.fit(x,y),feature_names=feature_column,filled=True,precision=3,proportion=True,rounded=True)
plt.show()
```
