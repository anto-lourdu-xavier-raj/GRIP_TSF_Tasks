```python
# Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Importing the dataset
data=pd.read_csv('SampleSuperstore.csv')
data
```

```python
data.head()
```

```python
# Import info
data.info()
```

```python
# Describe the dataset
data.describe()
```

```python
corr=data.corr()
corr
```

```python
sns.heatmap(corr, cmap='viridis',annot=True,xticklabels=corr.columns.values,yticklabels=corr.columns.values)
plt.show()
```

```python
data.columns
```

```python
# Check missing values
data.isnull().sum()
```

```python
# Check numerical features in column dataset
numerical_features = [feature for feature in data.columns if data[feature].dtypes !='O']
print('Number of Numerical Features :',len(numerical_features))
data[numerical_features].head()
```

```python
# Discrete features
discrete_features=[feature for feature in numerical_features if len(data[feature].unique())<25]
print('Number of Numerical Features :',len(discrete_features))
```

```python
['Quantity','Discount']

```

```python
discrete_features.append('Postal Code')
discrete_features
```

```python
# Count unique variable
for feature in discrete_features:
    sns.countplot(x=feature, data=data)
    plt.show()
```

```python
# Check for distribution
for feature in discrete_features:
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.show()
    
```

```python
# Relationship between discrete feature and profit
for feature in discrete_features:
    sns.barplot(x=feature,y='Profit',data=data)
    plt.xticks(rotation=60)
    plt.show()
```

    The relationship between Quantity and Profit is positive relation, So if the quantity increases the profit also get increase . As  we can see the relation between Discount and profit is negative. So with increase in Discount the profile getb decrease
    

```python
# Outliers
for feature in discrete_features:
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show
```

```python
# Continuos features
continuos_features=[feature for feature in numerical_features if feature not in discrete_features]
print('Number of continous features:',len(continuos_features))
```

```python
continuos_features
```

```python
continuos_features.remove('Profit')
continuos_features
```

```python
# Count the unique variables
for feature in continuos_features:
    sns.countplot(x=feature,data=data)
    plt.show()
```

```python
# Checking distribution
for feature in continuos_features:
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.show()
```

```python
# Relation between Continuos features and profit
for feature in continuos_features:
    plt.scatter(x=data[feature],y=data['Profit'],c='m')
    plt.title('Sales vs Profit')
    plt.xlabel('Sales')
    plt.ylabel('Profit')
    
```

```python
for feature in continuos_features:
    sns.barplot(x=feature,y='Profit',data=data)
    plt.xticks(rotation=60)
    plt.show()
```

```python
# Outliers
for feature in continuos_features:
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```

```python
# Categorical Variables
categorical_features=[feature for feature in data.columns if data[feature].dtypes=='O']
print('Number of categorical features:',len(categorical_features))
```

```python
data[categorical_features].head()
```

```python
for feature in categorical_features:
    print(f'The feature is ({feature})and the number of categories are ({len(data[feature].unique())})')
```

```python
for feature in categorical_features:
    sns.countplot(x=feature,data=data)
    plt.xticks(rotation=60)
    plt.show()
```

```python
#  Relationship between categorical features and profit
for feature in categorical_features:
    plt.figure(figsize=(11,7))
    sns.barplot(x=feature,y='Profit',data=data)
    plt.xticks(rotation=60)
    plt.show()
    
```

```python
# Pieplot in subcategory
plt.figure(figsize=(11,7))
data['Sub-Category'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

```python
# Total profit and sales in the basis of sub-category
plt.figure(figsize=(11,7))
data.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.show()
```

```python
# Profit VS Sales According to the region
sns.set_style('whitegrid')
sns.lmplot(x='Sales',y='Profit',col='Region',data=data,hue='Segment',palette='coolwarm')
```

    Here we can see the profit is highest in west region and then east region. In central Region the profit is increasing and decreasing as well

```python
# Quantity wise distribution of categories
data.groupby('Category')['Quantity'].sum().plot.pie(autopct='%1.1f%%')
plt.show()
```

```python
# Quantity wise distribution in regions
data.groupby('Region')['Quantity'].sum().plot.pie(autopct='%1.1f%%')
plt.show()
```

```python
# Top profitable cities
top_cities=data.groupby('City')['Profit'].sum().reset_index().sort_values(by='Profit',ascending=False)
top_cities
```

```python
sns.catplot('City','Profit',data=top_cities.head(10),kind='bar',height=6,aspect=3)
plt.show()
```

```python
# Top profitable sub-category
top_sub_categories=data.groupby('Sub-Category')['Profit'].sum().reset_index().sort_values(by='Profit',ascending=False)
top_sub_categories
```

```python
sns.catplot(x='Sub-Category',y='Profit',data=top_sub_categories.head(10),kind='bar',height=6,aspect=3)
top_sub_categories
```

```python
# Quantity wise distribution  of sub-category
data.groupby('Sub-Category')['Quantity'].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(11,7))
data.groupby('Sub-Category')['Quantity'].sum().sort_values(ascending=False).plot.pie(autopct='%1.1f%%')
plt.show()
```
